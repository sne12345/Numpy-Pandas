```python
matrix.groupby(["source","target"])["rating"].sum()
```

```python
matrix.groupby(["source","target"])["rating"].sum().unstack().fillna(0)
```

```python
def get_rating_matrix(filename):
    matrix = pd.read_csv(filename)
    return matrix.groupby(["source","target"])["rating"].sum().unstack().fillna(0).values
```

```python
get_rating_matrix("movie_rating.csv")
```

```python
def get_frequent_matrix(filename):
    freq = pd.read_csv(filename)
    freq["rating"] = 1
    return freq.groupby(["source", "target"])["rating"].sum().unstack().values
```

```python
get_frequent_matrix("1000i.csv")
```
