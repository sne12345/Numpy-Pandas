# Groupby

```python
import pandas as pd
```

```python
df = pd.read_csv('./PatientInfo.csv')
df.head()
```

```python
df.groupby("province")["contact_number"].sum()
```

```python
tmp = df.groupby(["province","sex"])["contact_number"].sum()
tmp
```

.

# unstack()

```python
tmp.unstack()
```

```python
tmp.swaplevel()
```

```python
tmp.sum(level=0)
```

```python
tmp.sum(level=1)
```

# 그룹을 기준으로 filtering

```python
df.groupby('province').filter(lambda x: len(x) < 15).head()
```

```python
df.groupby('city').filter(lambda x: x['contact_number'].sum() > 500)
```
