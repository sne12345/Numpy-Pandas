# Groupby

```python
import pandas as pd
```

```python
df = pd.read_csv('./PatientInfo.csv')
df.head()
```

```python
df.groupby("province")["contact_number"].sum()
```

```python
tmp = df.groupby(["province","sex"])["contact_number"].sum()
tmp
```

.

# unstack()

```python
tmp.unstack()
```

```python
tmp.swaplevel()
```

```python
tmp.sum(level=0)
```

```python
tmp.sum(level=1)
```

.

# grouped

```python
grouped = df.groupby("sex")
grouped
```

```python
for name, group in grouped:
    print(name)
    print(group)
```

```python
grouped.get_group("male").head()
```

```python
grouped.agg(sum)
```

```python
import numpy as np
grouped.agg(np.mean)
```

# 특정 컬럼에 여러 개의 Function apply

```python
grouped["contact_number"].agg([np.sum, np.mean, np.std])
```
