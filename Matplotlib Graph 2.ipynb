```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Image
from pandas.plotting import scatter_matrix
```

```python
Image('./boston.png')
```

```python
df = pd.read_excel('./BostonHousing.xls')
df
```

```python
fig = plt.figure()
ax = []
for i in range(1,5):
    ax.append(fig.add_subplot(2,2,i))
ax[0].scatter(df["CRIM"], df["MEDV"])
ax[1].scatter(df["LSTAT"], df["MEDV"])
ax[2].scatter(df["INDUS"], df["MEDV"])
ax[3].scatter(df["RM"], df["MEDV"])
plt.show()
```

# Scatter Matrix

```python
scatter_matrix(df, diagonal="kde", alpha=1, figsize=(20,20))
```

# 온도 그래프

```python
from sklearn.preprocessing import StandardScaler
std_scaler = StandardScaler()
scale_data = std_scaler.fit_transform(df)
scale_data
```

```python
corr_data = np.corrcoef(scale_data.T)
corr_data
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(corr_data, vmin=-1, vmax=1, interpolation="nearest")
fig.colorbar(cax)
fig.set_size_inches(10,10)
ticks=np.arange(0,15,1)
ax.set_xticks(ticks)
ax.set_yticks(ticks)

ax.set_xticklabels(df.columns)
ax.set_yticklabels(df.columns)
```
