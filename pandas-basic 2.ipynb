# Series Operation

### <span style="color:red">인덱스가 겹치는 부분만 연산을 실행한다.</span>

```python
s1 = Series(range(1,6), index = list("abcde"))
s1
```

# DataFrame Operation

### 리스트로 인덱스 설정하기

```python
df1 = DataFrame(np.arange(9).reshape(3,3), index=list("abc"))
df1
```

```python
df2 = DataFrame(np.arange(16).reshape(4,4), index=list("abcd"))
df2
```

```python
df1 + df2
```

```python
df1.add(df2, fill_value=0)
```

.

# Series + DataFrame

### 리스트로 컬럼명 설정하기

```python
list("abcd")
```

```python
df = DataFrame(np.arange(16).reshape(4,4), columns=list("abcd"))
df
```

```python
s = Series(np.arange(10,14), index=list("abcd"))
s
```

### <span style="color:red">column을 기준으로 broadcasting한다.</span>

```python
df + s
```
