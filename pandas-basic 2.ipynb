# DataFrame Operation

### 리스트로 인덱스 설정하기

```python
df1 = DataFrame(np.arange(9).reshape(3,3), index=list("abc"))
df1
```

```python
df2 = DataFrame(np.arange(16).reshape(4,4), index=list("abcd"))
df2
```

```python
df1 + df2
```

```python
df1.add(df2, fill_value=0)
```

# apply for dataframe

```python
df_info = df_age[["confirmed","deceased"]]
df_info
```

```python
f = lambda x : x.max() - x.min()
df_info.apply(f)
```

```python
def f(x):
    return Series([x.max(), x.min()], index=["max","min"])
df_info.apply(f)
```

# describe

```python
df_info.describe()
```

```python
f = lambda x : (int)(x)
df_info.describe().applymap(f)
```
